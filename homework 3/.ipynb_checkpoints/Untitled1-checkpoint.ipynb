```python
# Import necessary libraries and depencencies
import pandas as pd
import numpy as np
import datetime as dt
from pathlib import Path

%matplotlib inline
```

```python
# Set the file paths for the CSVs using the Path class from the pathlib library
whale_returns_path = Path('Resources/whale_returns.csv')
algo_returns_path = Path('Resources/algo_returns.csv')
sp500_history_path = Path('Resources/sp500_history.csv')
```

```python
# Reading whale returns
whale_returns_csv = Path("Resources/whale_returns.csv")
whale_df = pd.read_csv(whale_returns_csv,index_col= "Date",infer_datetime_format=True,parse_dates=True)
whale_df.sort_index(inplace=True)
whale_df.head()
```

```python
# Count nulls
whale_df.isnull().sum()
```

```python
# Drop nulls
whale_df.dropna(inplace =True)
whale_df.isnull().sum()
```

```python
whale_df.head()
```

```python
# Reading algorithimic returns
algo_returns_csv = Path("Resources/algo_returns.csv")
algo_df = pd.read_csv(algo_returns_csv,index_col= "Date",infer_datetime_format=True,parse_dates=True)
algo_df.sort_index(inplace=True)
algo_df.head()
```

```python
# Count nulls
algo_df.isnull().sum()
```

```python
# Drop nulls
algo_df.dropna(inplace = True)
algo_df.isnull().sum()
```

```python
algo_df.head()
```

```python
# Reading S&P 500 Closing Prices
sp500_history_csv = Path("Resources/sp500_history.csv")
sp500_df = pd.read_csv(sp500_history_csv,index_col='Date',infer_datetime_format=True,parse_dates=True)
sp500_df.sort_index(inplace = True)
sp500_df.head()
```

```python
# Check Data Types
sp500_df.dtypes
```

```python
# Fix Data Types
sp500_df['Close'] = sp500_df['Close'].str.replace('$','')
sp500_df['Close'] = sp500_df['Close'].astype('float')

sp500_df.dtypes
```

```python
# Calculate Daily Returns

sp500_daily_returns = sp500_df.pct_change()
sp500_daily_returns.head()
```

```python
# Drop nulls
sp500_daily_returns.dropna(inplace=True)
sp500_daily_returns
```

```python
# Rename `Close` Column to be specific to this portfolio.
sp500_daily_returns.columns = ['S&P 500']
sp500_daily_returns.head()
```

```python
# Join Whale Returns, Algorithimic Retrns, and the S&P, and the S&P 500 Returns into a single DataFrame with columns for each portfolio's returns. 
combined_daily_returns = pd.concat([whale_df,algo_df,sp500_daily_returns],axis="columns",join="inner")
combined_daily_returns.sort_index(inplace=True)
combined_daily_returns.head()
```

```python
# Plot daily returns of all portfolios
combined_daily_returns.plot(figsize=(50,20), title="Daily Returns of Portfolios and S&P500")
```

```python
# Calculate cumulative returns of all portfolios

# Plot cumulative returns
cumulative_returns = (1 + combined_daily_returns).cumprod() - 1
cumulative_returns.plot(figsize=(50,20), title="Cumulative Returns of the Portfolios and S&P500 ")


```

```python
# Box plot to visually show risk
cumulative_returns.plot.box(figsize=(50,20), grid=True, title="Cumulative return of porfolios in box " )
```

```python
# Calculate the daily standard deviations of all portfolios
all_portfolio_std = combined_daily_returns.std()
all_portfolio_std
```

```python
# Calculate the daily standard deviation of S&P 500
sp500_std = sp500_daily_returns.std()
sp500_std
```

```python
# Determine which portfolios are riskier than the S&P 500
all_portfolio_std = all_portfolio_std.sort_values(ascending=False)
all_portfolio_std
```

```python
# Calculate the annualized standard deviation (252 trading days)
annualized_std = all_portfolio_std * np.sqrt(252)
annualized_std
```

```python
# Calculate the rolling standard deviation for all portfolios using a 21-day window
ax = combined_daily_returns['S&P 500'].rolling(window=21).std().plot(figsize=(50,20),lw=8, color="Magenta")

# Plot the rolling standard deviation
```

```python
# Calculate the correlation
returns_correlations = combined_daily_returns.corr()
returns_correlations

# Display the correlation matrix
```
